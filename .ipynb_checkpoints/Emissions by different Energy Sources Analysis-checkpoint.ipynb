```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/GCB2022v27_MtCO2_flat.csv')
print(data.info())
```

```python
print(data.head())
```

```python
data.describe()
```

```python
data2= data[data['Country'] =='Albania']
print(data2)
```

```python
data.describe(include='all')
```

```python
plt.hist(data2['Coal'],color="blue")
plt.xlabel('Amount of Coal')
plt.ylabel("Frequency")
plt.title("Coal emissions",color="black")
plt.show()
```

```python
Conclusion: Very less countries use coal as their primary source of energy.
```

```python
plt.hist(data2['Oil'],color="black")
plt.xlabel('Amount of Oil')
plt.ylabel("Frequency")
plt.title("Oil emissions",color="black")
plt.show()
```

```python
Conclusion: Wider range of countries use oil as energy resources
```

```python
plt.hist(data2['Gas'],color="green")
plt.xlabel('Amount of Gas')
plt.ylabel("Frequency")
plt.title("Gas emissions",color="black")
plt.show()
```

```python
Conclusion: Very less countries use gas as their primary source of energy.
```

```python
plt.scatter(data2['Year'],data2['Coal'])
plt.title("Coal production in Albania", color='black')
plt.xlabel("Year",color="black")
plt.ylabel("Coal emmisions",color="black")
plt.show()
```

```python
Coal Production exponential increased in 1980s and exponentially decreased in late 90s during their economic downfall.
```

```python
plt.scatter(data2['Year'],data2['Oil'])
plt.plot(data2['Year'],data2['Oil'])
plt.title("Oil production in Albania", color='black')
plt.xlabel("Year",color="black")
plt.ylabel("Oil emmisions",color="black")
plt.show()
```

```python
plt.scatter(data2['Year'],data2['Gas'])
plt.title("Gas production in Albania over the years", color='black')
plt.xlabel("Year",color="black")
plt.ylabel("Gas emmisions",color="black")
plt.show()
```
