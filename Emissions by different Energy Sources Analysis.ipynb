```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/GCB2022v27_MtCO2_flat.csv')
print(data.info())
```

```python
print(data.head())
```

```python
data.describe()
```

```python
data2= data[data['Country'] =='Albania']
print(data2)
```

```python
data.describe(include='all')
```

```python
plt.hist(data2['Coal'],color="blue")
plt.xlabel('Amount of Coal')
plt.ylabel("Frequency")
plt.title("Coal emissions",color="black")
plt.show()
```

```python
Conclusion: Very less countries use coal as their primary source of energy.
```

```python
plt.hist(data2['Oil'],color="black")
plt.xlabel('Amount of Oil')
plt.ylabel("Frequency")
plt.title("Oil emissions",color="black")
plt.show()
```

```python
Conclusion: Wider range of countries use oil as energy resources
```

```python
plt.hist(data2['Gas'],color="green")
plt.xlabel('Amount of Gas')
plt.ylabel("Frequency")
plt.title("Gas emissions",color="black")
plt.show()
```

```python
Conclusion: Very less countries use gas as their primary source of energy.
```
